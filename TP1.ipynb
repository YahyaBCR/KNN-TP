```python
filename = r"C:\Users\yahya\OneDrive\Documents\emsi\DeepLearning\creditcard.csv"
```

```python
data = pd.read_csv(filename)
```

```python
print(data.shape)
print(data.head())
```

```python
y = data["Class"]
X = data[["V1","V2","V3","V4","V5","V6","V7","V8","V9","V10","V11","V12","V13","V14","V15","V16","V17","V18","V19","V20","V21","V22","V23","V24","V25","V26","V27","V28","Amount"]]
```

```python
X.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=4)
```

```python
print(X_train.shape)
print(y_train.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import metrics
```

```python
k_range = range(1, 26)
scores = {}
score_list = []
```

```python
for k in k_range:
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    
    y_predict = knn.predict(X_test)
    
    
    scores[k] = metrics.accuracy_score(y_test, y_predict)
    score_list.append(metrics.accuracy_score(y_test, y_predict))


    print("Taille de X_train:", X_train.shape)
    print("Taille de X_test:", X_test.shape)
```
